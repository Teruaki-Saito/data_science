```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
homo_means = [23, 35, 40, 12, 25]
unknown_means = [40, 10, 5, 50, 60]
special_means = [45, 54, 34, 12, 50]

x = np.arange(len(labels)) # the label location
width = 0.15 # the width of the bars

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 10), tight_layout=True)
axes = axes.flatten()
axes[0].bar(x - width/2, men_means, width, label='Men')
axes[0].bar(x + width/2, women_means, width, label='Women')
axes[0].set_xlabel('xlabel_1', fontsize=10)
axes[0].set_ylabel('ylabel_1', fontsize=10)
axes[0].set_title('title_1', fontsize=15)
axes[0].set_xticks(x) # 座標を取得している？
axes[0].set_xticklabels(labels) # set_xticksの上にラベルをセットしている？
axes[0].legend(loc='upper left')

axes[1].bar(x - width, men_means, width, label='Men')
axes[1].bar(x, women_means, width, label='Women')
axes[1].bar(x + width, homo_means, width, label='Homo')
axes[1].set_xlabel('xlabel_2', fontsize=10)
axes[1].set_ylabel('ylabel_2', fontsize=10)
axes[1].set_title('title_2', fontsize=15)
axes[1].set_xticks(x) # 座標を取得している？
axes[1].set_xticklabels(labels) # set_xticksの上にラベルをセットしている？
axes[1].legend(loc='upper right')

axes[2].bar(x - 1.5*width, men_means, width, label='Men')
axes[2].bar(x - 0.5*width, women_means, width, label='Women')
axes[2].bar(x + 0.5*width, homo_means, width, label='Homo')
axes[2].bar(x + 1.5*width, unknown_means, width, label='Unknown')
axes[2].set_xlabel('xlabel_3', fontsize=10)
axes[2].set_ylabel('ylabel_3', fontsize=10)
axes[2].set_title('title_3', fontsize=15)
axes[2].set_xticks(x) # 座標を取得している？
axes[2].set_xticklabels(labels) # set_xticksの上にラベルをセットしている？
axes[2].legend(loc='upper right')

axes[3].bar(x - 2*width, men_means, width, label='Men')
axes[3].bar(x - width, women_means, width, label='Women')
axes[3].bar(x, homo_means, width, label='Homo')
axes[3].bar(x + width, unknown_means, width, label='Unknown')
axes[3].bar(x + 2*width, special_means, width, label='Special')
axes[3].set_xlabel('xlabel_4', fontsize=10)
axes[3].set_ylabel('ylabel_4', fontsize=10)
axes[3].set_title('title_4', fontsize=15)
axes[3].set_xticks(x) # 座標を取得している？
axes[3].set_xticklabels(labels, rotation=320) # set_xticksの上にラベルをセットしている？
axes[3].legend(loc='upper right')

axes[4].bar(x - 1.5*width, men_means, width, label='Men')
axes[4].bar(x - 0.5*width, women_means, width, label='Women')
axes[4].bar(x + 0.5*width, homo_means, width, label='Homo')
axes[4].bar(x + 1.5*width, unknown_means, width, label='Unknown')
axes[4].plot(labels, special_means, label='special')
axes[4].set_xlabel('xlabel_4', fontsize=10)
axes[4].set_ylabel('ylabel_4', fontsize=10)
axes[4].set_title('title_4', fontsize=15)
axes[4].set_xticks(x) # 座標を取得している？
axes[4].set_xticklabels(labels, rotation=320) # set_xticksの上にラベルをセットしている？
axes[4].legend(loc='upper right')
;
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.15  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, men_means, width, label='Men')
rects2 = ax.bar(x, women_means, width, label='Women')
rects3 = ax.bar(x + width, homo_means, width, label='Homo')
ax.plot(labels, special_means, label='special')
ax.set_xlabel('xlabel_2', fontsize=10)
ax.set_ylabel('ylabel_2', fontsize=10)
ax.set_title('title_2', fontsize=15)
ax.set_xticks(x) # 座標を取得している？
ax.set_xticklabels(labels) # set_xticksの上にラベルをセットしている？
ax.legend(loc='upper right')

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
