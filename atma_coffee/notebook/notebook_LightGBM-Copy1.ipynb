```python
import pandas as pd
import numpy as np
import pandas_profiling as pdp
import sweetviz as sv

import os
import sys
from pygam import s, LinearGAM
import category_encoders as ce

import matplotlib.pyplot as plt
import seaborn as sns

import ptitprince
from matplotlib_venn import venn2
plt.rcParams['font.family'] = 'Hiragino Maru Gothic Pro'

from common_module.utils.plot import (
    plot_venn,
    plot_histogram,
    plot_raincloud,
    plot_heatmap,
    plot_histgram_for_valid,
    plot_lineargam,
    plot_feature_importance_for_valid
)
from common_module.utils.summarize import (
    get_pandas_profiling,
    get_sweetviz_report
)
from common_module.utils.preprocessor import (
    OrdinalEncodingBlock,
    CountEncodingBlock,
    GroupingBlock
)
from common_module.utils.format_df import format_df

from atma_coffee.modules.config_manager import ConfigManager
from atma_coffee.modules.features import (
    make_primary_key,
    get_world_country_features,
    get_altitude_features,
    get_num_bags_feature,
    get_bag_weight_features,
    get_hervest_year_features,
    get_grading_date_features,
    get_expiration_features,
    get_moisture_features,
    get_defects_features,
    get_cross_num_features,
    merge_features
)
from atma_coffee.modules.train_module import TrainLGBModule
from atma_coffee.modules.predict import make_pred_df
from atma_coffee.modules.validator import (
    make_stratifiedKFold,
    add_true_for_valid,
    calc_mean_absolute_error
)
from atma_coffee.modules.postprocessor import ceil_pred_cols

sys.path.append("../")
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 100)

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
config_path = "../config/setting.yaml"
config_manager = ConfigManager(config_path=config_path)
```

```python
# read
df_train_origin = pd.read_csv("../input/train.csv", encoding="utf-8-sig")
df_test_origin = pd.read_csv("../input/test.csv", encoding="utf-8-sig")
df_sub = pd.read_csv("../input/coffee-review_sample_submission.csv", encoding="utf-8-sig")

# shape
display(df_train_origin.shape)
display(df_test_origin.shape)
display(df_sub.shape)
```

```python
# train
# primary_keyを作成する
if "primary_key" not in df_train_origin.columns:
    df_train = make_primary_key(input=df_train_origin, is_train=True)
df_train, encoder_ceb = merge_features(input=df_train, is_train=True, config=config_manager)

# group feature
ins_group = GroupingBlock(
    group_keys=config_manager.group_encoding_cols["group_keys"],
    target_cols=config_manager.group_encoding_cols["target_cols"],
    methods=config_manager.group_encoding_cols["methods"]
)
df_agg_set, df_train = ins_group.fit_transform(input=df_train)
df_train = format_df(df_train, config_manager.feature_columns)

display(df_train.head(2))
display(df_train.shape)
```

```python
# test
# primary_keyを作成する
if "primary_key" not in df_test_origin.columns:
    df_test = make_primary_key(input=df_test_origin, is_train=False)
df_test = merge_features(input=df_test, is_train=False, config=config_manager, encoder_ceb=encoder_ceb)
df_test = ins_group.transform(input=df_test, df_agg_set=df_agg_set)
df_test = format_df(df_test, config_manager.feature_columns)

display(df_test.head(2))
```

### LightGBM

```python
model_set = {}
importance_set = {}
evals_set = {}
df_val_set = {}

# from sklearn.model_selection import train_test_split
# df_train_t, df_train_v = train_test_split(df_train, test_size=0.25, random_state=0)
# print(df_train_t.shape)
# print(df_train_v.shape)

for i, target in enumerate(config_manager.pred_cols):
    print(f"train {target} ...")
    # stratified kfolds
    fold_idx = make_stratifiedKFold(
        df_train,
        df_train[target],
        n_splits=config_manager.n_splits,
        random_state=config_manager.random_state
    )
    # train
    for cv_num, (tr_idx, va_idx) in enumerate(fold_idx):
        df_train_t, df_train_v = df_train.loc[tr_idx, :], df_train.loc[va_idx, :]
        ins_train = TrainLGBModule(config=config_manager, target=target)
        trained_model, df_imp, evals_result = ins_train.train_model(df_train=df_train_t, df_val=df_train_v, have_weight=False)
        # dataset への保存
        model_set[f"{target}_{cv_num}"] = trained_model
        importance_set[f"{target}_{cv_num}"] = df_imp
        evals_set[f"{target}_{cv_num}"] = evals_result
        # validationデータを予測
        df_val_pred = make_pred_df(model=trained_model, target=target, df_test=df_train_v, config=config_manager, is_test=False, cv_num=cv_num)
        df_val = add_true_for_valid(df_val=df_train_v, df_val_pred=df_val_pred, target=target, cv_num=cv_num, config=config_manager)
        mae_score = calc_mean_absolute_error(
            true=df_val[df_val["cv_num"] == cv_num][f"true_{target}"],
            pred=df_val[df_val["cv_num"] == cv_num][f"pred_{target}"])
        print(f"■ target_col={target} | cv_num={cv_num} | mae_score={mae_score:.4f}")
        df_val_set[f"{target}_{cv_num}"] = df_val

    print("=" * 100)
```

```python
target = "total_cup_points"
for n in range(config_manager.n_splits):
    plot_histgram_for_valid(
        data_true=df_val_set[f"{target}_{n}"][f"true_{target}"],
        data_pred=df_val_set[f"{target}_{n}"][f"pred_{target}"],
        title=f"target={target} | cv_num={n}",
        figsize=(3,3)
    )
```

```python
plot_feature_importance_for_valid(importance_set, target="total_cup_points", n_splits=config_manager.n_splits)
```

```python
df_pred_cv_set = {}
df_pred_target_set = {}

for target in config_manager.pred_cols:
    dfs = []
    for cv_num in range(config_manager.n_splits):
        print(f"■ target_col={target} | cv_num={cv_num}")
        model = model_set[f"{target}_{cv_num}"]
        df_pred_part = make_pred_df(model=model, target=target, df_test=df_test, config=config_manager, is_test=True, cv_num=cv_num)
        df_pred_cv_set[f"{target}_{cv_num}"] = df_pred_part
        dfs.append(df_pred_part)
    df_pred_target = pd.concat(dfs, ignore_index=True, sort=False)
    df_pred_target = df_pred_target.rename(columns={f"pred_{target}": target})
    df_pred_target_set[target] = df_pred_target.groupby(config_manager.primary_key, as_index=False, sort=False)[target].mean()  # defaultでソートされるため
    print("=" * 100)
```

```python
for i, target in enumerate(config_manager.pred_cols):
    if i == 0:
        df_submission = df_pred_target_set[target]
    else:
        df_submission = pd.merge(df_submission, df_pred_target_set[target], how="left", on=config_manager.primary_key)

display(df_submission.head(2))
```

```python
df_submission = ceil_pred_cols(
    input=df_submission,
    target_cols=config_manager.one_hundred_pred_cols,
    ceil_value=config_manager.ceil_value
)

display(df_submission.head(2))
```

```python
suffix = "5th_add_agg_ceil"
df_submission_ = df_submission.drop("primary_key", axis=1)
df_submission_.to_csv(f"../output/df_submission_{suffix}.csv", index=False)
```
